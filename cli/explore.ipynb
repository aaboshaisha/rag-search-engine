```python
def search(qry:str) -> list[str]:
    matches, count = [], 0
    for m in movies:
        if qry in m['title']:
            count += 1
            print(f'{count}. {m["title"]}')
            matches.append(m)
    res = [m['title'] for m in sorted(matches, key=lambda x: x['id'])][:5]
    return res
```
